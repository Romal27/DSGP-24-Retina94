```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load DialoGPT model and tokenizer
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Set padding token explicitly
tokenizer.pad_token = tokenizer.eos_token

# Define fallback responses
def generate_safe_response(user_input):
    if "diabetic retinopathy" in user_input.lower():
        return "Diabetic retinopathy is a condition where high blood sugar damages the retina. Proper management, such as controlling blood sugar and regular eye check-ups, can help prevent complications."
    return None

# Chatbot conversation loop
chat_history_ids = None
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Bot: Goodbye!")
        break

    # Check for fallback response
    safe_response = generate_safe_response(user_input)
    if safe_response:
        print(f"Bot: {safe_response}")
        continue

    # Tokenize user input
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors="pt")
    attention_mask = torch.ones(new_input_ids.shape, dtype=torch.long)  # Explicit attention mask

    # Combine with chat history
    if chat_history_ids is not None:
        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
        attention_mask = torch.cat([torch.ones_like(chat_history_ids), attention_mask], dim=-1)
    else:
        bot_input_ids = new_input_ids

    # Generate response with sampling
    output = model.generate(
        bot_input_ids,
        attention_mask=attention_mask,  # Pass the attention mask
        max_length=100,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,   # Enable sampling for diverse responses
        top_k=50,         # Restrict sampling to top 50 probable tokens
        top_p=0.9,        # Use nucleus sampling for coherent results
    )
    # Decode the response
    response = tokenizer.decode(output[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    print(f"Bot: {response}")

    # Update chat history
    chat_history_ids = output
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load BioGPT
tokenizer = AutoTokenizer.from_pretrained("microsoft/BioGPT")
model = AutoModelForCausalLM.from_pretrained("microsoft/BioGPT")

# Chatbot logic
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Bot: Goodbye!")
        break

    # Add structured prompt to guide the model
    prompt = f"This is a medical chatbot specialized in diabetic retinopathy. Answer the following question: {user_input}"
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate response with controlled sampling
    output = model.generate(
        inputs["input_ids"],
        max_length=100,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id,
    )
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    print(f"Bot: {response}")
```

```python
# Rule-based responses
def get_knowledge_base_response(user_input):
    knowledge_base = {
        "what is diabetic retinopathy": "Diabetic retinopathy is a condition where high blood sugar damages the retina.",
        "how can I manage it": "You can manage it by controlling blood sugar levels, regular eye exams, and maintaining a healthy lifestyle.",
    }
    for key, response in knowledge_base.items():
        if key.lower() in user_input.lower():
            return response
    return "I'm sorry, I didn't understand that. Can you rephrase?"

# Chatbot logic
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Bot: Goodbye!")
        break
    response = get_knowledge_base_response(user_input)
    print(f"Bot: {response}")
```

```python
# Combined logic with fallback to a model
from transformers import pipeline

# Load a generic Q&A model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Knowledge base
def get_knowledge_base_response(user_input):
    knowledge_base = {
        "what is diabetic retinopathy": "Diabetic retinopathy is a condition where high blood sugar damages the retina.",
        "how can I manage it": "You can manage it by controlling blood sugar levels, regular eye exams, and maintaining a healthy lifestyle.",
    }
    for key, response in knowledge_base.items():
        if key.lower() in user_input.lower():
            return response
    return None

# Chatbot logic
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Bot: Goodbye!")
        break

    # Rule-based response
    knowledge_response = get_knowledge_base_response(user_input)
    if knowledge_response:
        print(f"Bot: {knowledge_response}")
        continue

    # Fallback to the model
    context = (
        "Diabetic retinopathy is a complication of diabetes that affects the eyes. "
        "It is caused by damage to the blood vessels in the retina and can lead to blindness."
    )
    response = qa_pipeline(question=user_input, context=context)
    print(f"Bot: {response['answer']}")
```
